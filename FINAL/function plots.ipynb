```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x,a=1):
    return 1-np.exp(-x/a)

def g(z,x):
    return 1-f(x)*np.exp(z/1)
```

```python
x = np.linspace(-1,5,200)
z = np.linspace(-10,0,200)
```

```python
plt.plot(x,f(x))
```

```python
plt.plot(g(z,0),z)
plt.xlim(0,2)
```

```python
plt.plot(x,1-np.exp(-x))
```

```python
plt.plot(x,np.log(x))
```
