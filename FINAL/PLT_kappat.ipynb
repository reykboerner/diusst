```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def exp(z,a,b):
    return (1-b*np.exp(z/a))/(1-b*np.exp(-10/a))

xx = np.linspace(-10,0,200)
```

```python
plt.rcParams['mathtext.fontset']= 'cm'
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

fig, ax = plt.subplots(1,4,figsize=(6,1.8),dpi=150,sharey=True)

ax[0].plot(np.ones(len(xx)),xx,c='dodgerblue')
ax[1].plot(-xx/10,xx,c='dodgerblue')
ax[2].plot(exp(xx,1,1),xx,c='dodgerblue')
ax[2].plot(exp(xx,3,0.5),xx,c='firebrick')
ax[2].plot(exp(xx,3,1),xx,c='navy')

ax[3].plot(exp(xx,2,0),xx,c='dodgerblue')
ax[3].plot(exp(xx,2,0.2),xx,c='mediumseagreen')
ax[3].plot(exp(xx,2,0.5),xx,c='orange')
ax[3].plot(exp(xx,2,1),xx,c='firebrick')
ax[3].text(0.1,-2.5,'S=1',c='firebrick')
ax[3].text(1.1,-2.,'S=0',c='dodgerblue')
ax[0].set(ylabel='Depth (m)')

titles=['BASE','LIN','EXP','STAB']
for i in range(4):
    ax[i].set(xlim=(0,1.3),ylim=(-10,1),xticks=[0,1],xticklabels=['0',r'$\tilde \kappa_0$'])
    ax[i].set_title(titles[i],fontsize=8)
    ax[i].spines['right'].set_color('none')
    ax[i].spines['top'].set_color('none')
    ax[i].axhline(y=0,ls='--',c='k',lw=1)
    #ax[i].plot(1.1, -10, ">k", transform=ax[i].get_yaxis_transform(), clip_on=False)
    #ax[i].plot(0, 1, "^k", transform=ax[i].get_xaxis_transform(), clip_on=False)
    
plt.tight_layout(w_pad=1)
#plt.savefig('../../Masterarbeit/thesis/figs/diffu_versions.pdf', bbox_inches='tight')

plt.show()
```
