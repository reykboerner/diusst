```python
param_names = ['kappa','mu','attenu']
#name1 = '../output_files/220131'
name1 = '../code/bayesian/220131-202129_PG_EXP_5p-A1.h5'
#parameter that excludes 0 in range
param_nonzero = 2
finish = False
```
